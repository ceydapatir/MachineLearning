```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import pandas as pd
```

```python
np.random.seed(3)
factor=np.random.normal(0.5,1,(75,2))
x1=[0.8,0.8]+factor
x2=factor

X = np.r_[x1,x2]
y1 = np.zeros((75))
y2 = np.ones((75))


y = np.r_[y1,y2]
```

```python
C_range = list(range(1, 50,2))
G_range = ['auto', 'scale']
SVM_scores = []
for c in C_range:
    for g in G_range:
        SVModel=SVC( C=c, gamma=g).fit(X,y)
        scores = cross_val_score(SVModel, X, y, cv=10, scoring='accuracy')
        SVM_scores.append([scores.mean(),c,g])
print(SVM_scores)
```

```python
pd.DataFrame(SVM_scores)
```

```python
param={
    'C':np.arange(1,50,2),
    'gamma':('auto', 'scale')
}
```

```python
SVModel=SVC()
GridS=GridSearchCV(SVModel,param,cv=10).fit(X,y)
GridS.best_params_
```

```python
pd.DataFrame(GridS.cv_results_)[['mean_test_score', 'params']]
```
